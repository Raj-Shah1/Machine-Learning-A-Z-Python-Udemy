```
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Importing the dataset
datasets=pd.read_csv('gdrive/My Drive/Data/Churn_Modelling.csv')
X=datasets.iloc[:, 3:13].values
y=datasets.iloc[:,13].values
```

```
#One Hot Encoding using Column Transformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer

#LabelEncode Gender Column 
labelencoder_X = LabelEncoder()
X[:, 2] = labelencoder_X.fit_transform(X[:, 2])

# Country column, remainder = 'passthrough' lets other columns pass through
ct = ColumnTransformer([("Country", OneHotEncoder(), [1])], remainder = 'passthrough') 
X = ct.fit_transform(X)
X = X[:, 1:]
```

```
#Split dataset into train-test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```
#FeatureScaling
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```
#Importing keras
import keras
from keras.layers import Dense
from keras import Sequential
```

```
#Initializing the ANN
classifier =  Sequential()

#Adding input layer and hidden layer
classifier.add(Dense(units = 6, init = 'uniform', activation = 'relu', input_dim = 11 ))

#Adding second hidden layer
classifier.add(Dense(units = 6 , init = 'uniform', activation = 'relu'))

#Adding the output layer
classifier.add(Dense(units = 1 , init = 'uniform', activation = 'sigmoid'))

#Compiling the ANN
#loss - To calculate the error between actual and predicted value
#optimizer - To update the values of weights and biases
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])

#Fitting the ANN to the training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```
y_pred = classifier.predict(X_test)

#Convert y_pred from probabilities to True/False
y_pred = (y_pred > 0.5)
```

```
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
