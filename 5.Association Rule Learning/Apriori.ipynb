```
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

datasets=pd.read_csv('gdrive/My Drive/Data/Market_Basket_Optimisation.csv', header = None)
```

```
transactions = []
for i in range(0, 7501):
  transactions.append([str(datasets.values[i,j]) for j in range(0, 20)])
```

```
print(transactions[4:5])
#print(datasets[0:5][:])
```

```
#Training apriori on the dataset
from apyori import apriori
```

```
#support = transactions containing an item M1/total transactions
#confidence(M1 -> M2) = transactions containing M1 and M2/transactions containing M1
#lift(M1 -> M2) = confidence(M1->M2)/support(m2)
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```
#Visualizing the results
results = list(rules)
```

```
print(pd.DataFrame(results))
```
